```python
# Using Pretrained Embeddings:
# Training on 5-6 sentences is useless
# Real predictions will come from pretrained embeddings

# Popular pretrained models:
# 1. Google News Word2Vec
# 2. GloVe
```

```python
import gensim.downloader as api

print("Loading pretrained model:")
# model = api.load("word2vec-google-news-300")
model = api.load("glove-wiki-gigaword-100")
# Common pretrained vectors:
# glove-wiki-gigaword-50
# glove-wiki-gigaword-100
# glove-wiki-gigaword-300
# glove-twitter-200
```

```python
print("similarity between two words:")
model.similarity("car","bus")

print("Top 5 similar words")
model.most_similar("machine", topn=5)
```
